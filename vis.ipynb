```python
import igor.binarywave as bw
import numpy as np
import matplotlib.pyplot as plt

# .ibw
def read_ibw(file_path):
    bw_data = bw.load(file_path)
    data = bw_data['wave']['wData']  
    return data

# 
def plot_data(data):
    plt.imshow(data.T, cmap='viridis', origin='lower')
    plt.colorbar()
    plt.title('SHG')
    plt.show()


file_path = '/Users/ootahiroki/Downloads/shg.ibw'  
data = read_ibw(file_path)

# 
data_copy = np.copy(data)

# xx
data_copy[data_copy > 10000] = 10000

# 2
if data_copy.ndim == 2:
    plot_data(data_copy)
else:
    print(f"{data_copy.ndim}2")
```
